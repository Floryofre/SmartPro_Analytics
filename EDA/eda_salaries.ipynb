```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Cargar los archivos CSV
df_salary = pd.read_csv('data/nba_salaries_db.csv')
```

```python
# Establecer el estilo de seaborn para mejorar la estética
sns.set(style="whitegrid")

# Crear el histograma
plt.figure(figsize=(10, 6))
sns.histplot(df_salary['salary_2022_23'], bins=20, kde=True, color="skyblue", edgecolor='black', linewidth=1.2)

# Calcular el salario medio y mediano
mean_salary = df_salary['salary_2022_23'].mean()
median_salary = df_salary['salary_2022_23'].median()

# Añadir líneas verticales para el salario medio y mediano
plt.axvline(mean_salary, color='r', linestyle='--', linewidth=2)
plt.axvline(median_salary, color='g', linestyle='-', linewidth=2)

# Ajustar la posición de las anotaciones para mejorar la legibilidad
# Aumentar el valor de x para mover las anotaciones más a la derecha
right_edge = plt.xlim()[1]  # Obtener el límite derecho del eje x
offset = (right_edge - mean_salary) * 0.05  # Calcular un pequeño desplazamiento hacia la derecha

# Añadir anotaciones para el salario medio y mediano con la posición ajustada
plt.text(mean_salary + offset, plt.ylim()[1]*0.95, 'Media: ${:.2f}'.format(mean_salary), color='r')
plt.text(median_salary + offset, plt.ylim()[1]*0.9, 'Mediana: ${:.2f}'.format(median_salary), color='g')

# Añadir título y etiquetas con tamaños de fuente mayores
plt.title('Distribución de los salarios 2022-23', fontsize=18)
plt.xlabel('Salarios', fontsize=14)
plt.ylabel('Cantidad de jugadores', fontsize=14)

# Mostrar el gráfico
plt.show()
```

```python
# Establecer el estilo de seaborn para mejorar la estética
sns.set(style="whitegrid")

# Crear el boxplot de los salarios en orientación vertical
plt.figure(figsize=(10, 10))
box_plot = sns.boxplot(y=df_salary['salary_2022_23'], color='skyblue', linewidth=2.5)

# Añadir título y etiquetas con tamaños de fuente mayores para mejorar la legibilidad
box_plot.set_title('Distribución de los salarios 2022-23', fontsize=20)
box_plot.set_ylabel('Salarios', fontsize=16)

# Calcular estadísticas clave
median_salary = df_salary['salary_2022_23'].median()
q1_salary = df_salary['salary_2022_23'].quantile(0.25)
q3_salary = df_salary['salary_2022_23'].quantile(0.75)

# Añadir anotaciones para estadísticas clave

# Ajustar las posiciones de las anotaciones para la orientación vertical y desplazarlas hacia arriba
offset = 0.02 * (df_salary['salary_2022_23'].max() - df_salary['salary_2022_23'].min())  # Calcula un desplazamiento basado en el rango de los datos

plt.text(0, median_salary + offset, ' Mediana: ${:.2f}'.format(median_salary), ha='left', va='center', color='red', fontsize=12)
plt.text(0, q1_salary + offset, ' Q1: ${:.2f}'.format(q1_salary), ha='left', va='center', color='green', fontsize=12)
plt.text(0, q3_salary + offset, ' Q3: ${:.2f}'.format(q3_salary), ha='left', va='center', color='blue', fontsize=12)

# Mostrar la gráfica
plt.show()
```

Rendimiento en la cancha:

* Puntos anotados (PTS): Los jugadores que anotan muchos puntos suelen tener un mayor impacto en el juego y, por lo tanto, pueden ser más atractivos para los equipos.
* Asistencias (AST): Los jugadores con altos promedios de asistencias indican una buena visión de juego y habilidades de liderazgo.
* Rebotes (TRB): Los jugadores que dominan los rebotes, tanto ofensivos como defensivos, son valiosos para cualquier equipo.
* Robos (STL) y bloqueos (BLK): Un alto número de robos y bloqueos muestra habilidades defensivas excepcionales.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Configuración de los gráficos con un estilo más profesional
sns.set(style="whitegrid")
plt.figure(figsize=(20, 12))

# Definir una paleta de colores
palette = sns.color_palette("husl", 5)

# 1. Relación entre salario y puntos anotados por partido (PTS)
plt.subplot(2, 3, 1)
sns.regplot(x='salary_2022_23', y='PTS', data=df_salary, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'}, color=palette[0])
plt.xlabel('Salario ($)', fontsize=12)
plt.ylabel('Puntos anotados por partido', fontsize=12)
plt.title('Salario vs Puntos anotados por partido', fontsize=14)

# 2. Relación entre salario y minutos jugados por partido (MP)
plt.subplot(2, 3, 2)
sns.regplot(x='salary_2022_23', y='MP', data=df_salary, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'}, color=palette[1])
plt.xlabel('Salario ($)', fontsize=12)
plt.ylabel('Minutos jugados por partido', fontsize=12)
plt.title('Salario vs Minutos jugados por partido', fontsize=14)

# 3. Relación entre salario y asistencias por partido (AST)
plt.subplot(2, 3, 3)
sns.regplot(x='salary_2022_23', y='AST', data=df_salary, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'}, color=palette[2])
plt.xlabel('Salario ($)', fontsize=12)
plt.ylabel('Asistencias por partido', fontsize=12)
plt.title('Salario vs Asistencias por partido', fontsize=14)

# 4. Relación entre salario y rebotes totales por partido (TRB)
plt.subplot(2, 3, 4)
sns.regplot(x='salary_2022_23', y='TRB', data=df_salary, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'}, color=palette[3])
plt.xlabel('Salario ($)', fontsize=12)
plt.ylabel('Rebotes totales por partido', fontsize=12)
plt.title('Salario vs Rebotes totales por partido', fontsize=14)

# 5. Relación entre salario y robos de balón por partido (STL)
plt.subplot(2, 3, 5)
sns.regplot(x='salary_2022_23', y='STL', data=df_salary, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'}, color=palette[4])
plt.xlabel('Salario ($)', fontsize=12)
plt.ylabel('Robos de balón por partido', fontsize=12)
plt.title('Salario vs Robos de balón por partido', fontsize=14)

# 6. Relación entre salario y Triples anotados por partido
plt.subplot(2, 3, 6)
sns.regplot(x='salary_2022_23', y='3P', data=df_salary, scatter_kws={'alpha':0.6}, line_kws={'color': 'red'}, color=palette[0])
plt.xlabel('Salario ($)', fontsize=12)
plt.ylabel('Triples anotados por partido', fontsize=12)
plt.title('Salario vs Triples anotados por partido', fontsize=14)

# Ajustar el layout y mostrar el gráfico
plt.tight_layout()
plt.show()
```

Salario Promedio por posición en el juego

```python
#Base (PG): Director del juego, pasa y organiza.
#Escolta (SG): Anotador, especialmente buenos en tiros de larga distancia.
#Alero (SF): Versátil, puede hacer de todo un poco.
#Ala-pívot (PF): Fuerte, juega cerca del aro y recoge rebotes.
#Pívot (C): Más alto y fuerte, se queda cerca del aro para anotar y defender.

# Filtrar las posiciones relevantes y calcular el salario promedio por posición
positions = ['PG', 'SG', 'SF', 'PF', 'C']
average_salary_by_position = df_salary[df_salary['position'].isin(positions)].groupby('position')['salary_2022_23'].mean().reset_index()

# Crear el gráfico de barras para mostrar el salario promedio por posición
plt.figure(figsize=(10, 6))
sns.barplot(x='position', y='salary_2022_23', data=average_salary_by_position, palette='viridis')
plt.xlabel('Posición')
plt.ylabel('Salario Promedio ($)')
plt.title('Salario Promedio por Posición en la NBA')
plt.show()
```

Consistencia y durabilidad:

Partidos jugados (GP) y minutos jugados (MP): Los jugadores que juegan regularmente y tienen altos promedios de minutos por partido son más fiables y consistentes.

```python
# Crear una paleta de colores invertida
# Crear rangos de salario
salary_bins = [0, 5000000, 10000000, 15000000, 20000000, 25000000, 30000000, 35000000, 40000000, 45000000, float('inf')]
salary_labels = ['0-5M', '5-10M', '10-15M', '15-20M', '20-25M', '25-30M', '30-35M', '35-40M', '40-45M', '>45M']
df_salary['salary_range'] = pd.cut(df_salary['salary_2022_23'], bins=salary_bins, labels=salary_labels, right=False)

# Ordenar el DataFrame por 'MP' en orden descendente
top_players = df_salary.sort_values(by='salary_2022_23', ascending=False).head(5)

palette = sns.color_palette("Spectral", len(salary_labels))[::-1]

# Actualizar el gráfico de dispersión para usar la paleta de colores invertida
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GP', y='MP', data=df_salary, alpha=0.5, hue='salary_range', palette=palette)
plt.xlabel('Partidos Jugados (GP)')
plt.ylabel('Minutos Jugados por Partido (MP)')
plt.title('Consistencia y Durabilidad: Partidos Jugados vs Minutos Jugados por Partido - Rango de Salario')
plt.legend(title='Rango de Salario', bbox_to_anchor=(1, 1))

# Añadir anotaciones para los top 5 jugadores
for index, row in top_players.iterrows():
    plt.text(row['GP'], row['MP'], row['full_name'], color='black', fontsize=8)

plt.show()
```

Edad y proyección de carrera:

* Edad: Los jugadores más jóvenes tienen más años de carrera por delante y, por lo tanto, más oportunidades para mejorar y aumentar su valor.
* Proyección de desarrollo: Evaluar el potencial de desarrollo del jugador basado en su rendimiento pasado y sus habilidades actuales.

```python
# Filtrar jugadores que ganen menos de $3,000,000
df_filtered = df_salary[df_salary['salary_2022_23'] < 3000000]

# Crear una nueva columna para categorizar los PTS en grupos de 5 en 5 en el DataFrame filtrado
df_filtered['PTS_category'] = pd.cut(df_filtered['PTS'], bins=range(0, int(df_filtered['PTS'].max()) + 5, 5), right=False)

# Crear el gráfico de dispersión para Edad vs Salario, con el tamaño del punto representando los puntos anotados por partido (PTS) en el DataFrame filtrado
plt.figure(figsize=(12, 8))
sns.scatterplot(x='age', y='salary_2022_23', size='PTS', data=df_filtered, alpha=0.6, sizes=(20, 200), hue='PTS_category', palette='coolwarm')
plt.xlabel('Edad')
plt.ylabel('Salario ($)')
plt.title('Relación entre Edad y Salario de los Jugadores de la NBA, Categorizados por PTS (Salario < $3M)')
plt.legend(title='PTS Categoría', bbox_to_anchor=(1.05, 1), loc='upper left')

# Filtrar jugadores con más de 20 PTS en promedio en el DataFrame filtrado
top_scorers_filtered = df_filtered[df_filtered['PTS'] > 15]

# Añadir anotaciones para cada jugador con más de 20 PTS en el DataFrame filtrado
for index, row in top_scorers_filtered.iterrows():
    plt.text(row['age'], row['salary_2022_23'], row['full_name'], color='black', fontsize=9)

plt.show()
```

Entre los jugadores que anotan más de 15 puntos por partido, menores de 25 años y que ganan menos de medio millon de dolares al año, tenemos a Louis King, RaiQuan Gray y Skylar Mays, que destacan claramente por el gran potencial que tienen de crecer en su carrera profesional, lo que en 5 o 10 años más su valor en el mercado se podria multiplicar
